```python
import pandas as pd 
import sys

sys.path.append('../')


import utils.llm 


class Config:
    
    TARGET_QUIZ_RANGE = [i for i in range(1, 201)]
    DATA_PATH = '../data/'
    TEMPERATURE = 0.01

df = pd.read_csv(Config.DATA_PATH + 'df_JAQKET.csv')
```

```python
llm = utils.llm.LLM()
```

```python
from utils.scoring import sentence_splitter
from pprint import pprint
def check(query:str,response:str):
    
    queries = []
    outputs = []
    
    response_sentences = sentence_splitter(response)
    
    for ri in response_sentences:
        q = f"""「{query}」という質問に対して「{response}」という回答を得ました。このとき、回答文に含まれる「{ri}」という文は、正しいでしょうか？正しいのであれば、Yesを、正しくないのであれば、Noを出力してください。"""
        output = llm.generate(q,temperture=Config.TEMPERATURE)
        outputs.append(output)
        queries.append(q)
    
    
    pprint(outputs)
    
    return outputs,queries
    
```

```python

df_out = pd.DataFrame(columns=['qid','sentence_id''query','output'])




for i in Config.TARGET_QUIZ_RANGE:
    
    # show progress
    if i % 10 == 0:
        pprint("now processing the quiz of index... {}".format(i))
    
    qid = df.iloc[i]['qid']
    query = df.iloc[i]['question']
    response = df.iloc[i]['generated_answer']
    
    
    print (query)
    print (response)
    
    outputs,queries = check(query,response)
    
    for j in range(len(outputs)):
        
        #  The frame.append method is deprecated and will be removed from pandas in a future version. Use pandas.concat instead.
        df_out = pd.concat([df_out,pd.DataFrame([[qid,j,queries[j],outputs[j]]],columns=['qid','sentence_id','query','output'])])

df
```

```python
from datetime import datetime 
now = datetime.now()

df_out.to_csv(f'../data/df_out_benchmark_{now.strftime("%Y%m%d%H%M%S")}.csv',index=False)
```
