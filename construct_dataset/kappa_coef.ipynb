```python
import pandas as pd 
import numpy as np

df1 = pd.read_csv('./df_JAQKET_qa_annot1.csv')
df2 = pd.read_csv('./df_JAQKET_qa_annot2.csv')
```

```python
labels_annot1 = []
labels_annot2 = []
for i in range(len(df1)):
#     labels_ = df["factuality"][i].split(",")
#     labels.extend(labels_)

    lebals_annot1_ = df1["factuality"][i].split(",")
    lebals_annot2_ = df2["factuality"][i].split(",")
    
    labels_annot1.extend(lebals_annot1_)
    labels_annot2.extend(lebals_annot2_)


# labels = np.fromiter(labels, dtype=int)

labels_annot1 = np.fromiter(labels_annot1, dtype=int)
labels_annot2 = np.fromiter(labels_annot2, dtype=int)
```

```python
# カッパ係数を計算
from sklearn.metrics import cohen_kappa_score
kappa = cohen_kappa_score(labels_annot1, labels_annot2)
kappa
```
